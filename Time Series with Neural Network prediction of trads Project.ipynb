# Get stock opening price

```python
trd = A[["Open"]].values
```

```python
import numpy as np
np.array(A[["Open"]])
```

# MinmaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit_transform(trd)
```

# Create Data for training and testing

```python
trd.shape
```

```python
xtrain = []
ytrain = []
for i in range(50,1258,1):
    xtrain.append(trd[i-50:i])
    ytrain.append(trd[i:i+1])  
```

```python
type(xtrain)
```

# Conver data in Array

```python
xtr = np.array(xtrain)
```

```python
xtr.shape
```

```python
ytr = np.array(ytrain)
```

```python
ytr.shape
```

# Create a NN Model

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Dropout
```

```python
nn = Sequential()
nn.add(LSTM(50,return_sequences=True,input_shape=(50,1)))
nn.add(LSTM(100,return_sequences=True))
nn.add(Dense(1))
```

```python
nn.compile(optimizer="adam",loss="mean_squared_error")
nn.fit(xtr,ytr,epochs=10)
```
